# Fungsi

- hitung luas jika dual_sim = 1, jika tidak ada maka hitung kurang!
- luas = tinggi (px_height) * lebar (px_width)
- kurang = tinggi - lebar

```python
def luas (tinggi, lebar):
    hasil = tinggi * lebar
    return hasil

def kurang (tinggi, lebar):
    hasilk = tinggi - lebar
    return hasilk
```

```python
tr_new = tr.copy()

lk = [] #untuk memasukkan hasil luas dan kurang ke dalam list
j = 0 #index baris
#untuk hasil 


for i in tr_new['dual_sim'] :
    if i==1:
        lk.append(luas(tr_new['px_height'][j], tr_new['px_width'][j]))
        j += 1
    else:
        lk.append(kurang(tr_new['px_height'][j], tr_new['px_width'][j]))
    print(tr_new['px_height'][j])
    print(tr_new['px_width'][j])
tr_new['luas or kurang'] = lk #bikin kolom 'luas or kurang' dengan nilai dari list 'lk'
```
