### Querying Data

ini pokoknya buat data yang spesifik. gabisa pakai loc

```python
ds.head()
```

```python
ds[
    ds['year_id'] > 1950
]
```

kalo kondisinya lebih dari 1, tiap kondisinya harus dibungkus dengan ( )

```python
ds[
    (ds['year_id'] > 1950) &
    (ds['fran_id'] == 'Knicks') &
    (ds['pts'] > 150)
    ]
```

```python
ds[
    (ds['team_id'] == 'BLB' ) &
    (ds['pts'] > 100) &
    (ds['opp_pts'] > 100) &
    (ds['year_id'] == 1953)
]
```

cara kedua untuk query, yaitu pakai keyword **QUERY**

```python
ds.query( "team_id == 'BLB' & pts > 100 & opp_pts > 100 & year_id == 1953")
```

### Grouping and Aggregating Data

*self-explanatory* buat ngegroup data

```python
ds.groupby('fran_id')['pts'].sum().head(10)
```

```python
ds.groupby('fran_id')['pts'].max().head(10)
```

```python
ds.groupby('team_id')['pts'].max().head(10)
```

```python
ds[
    (ds['year_id'] > 2010) &
    (ds['fran_id'] == 'Knicks')
].groupby('year_id')['pts'].sum()
```

```python
ds[
    (ds['year_id'] > 2010) &
    (ds['fran_id'] == 'Knicks')
].groupby(['year_id', 'game_location'])['pts'].sum()
```

```python
ds[
    (ds['year_id'] > 2010) &
    (ds['fran_id'] == 'Knicks')
].groupby(['year_id', 'team_id', 'game_location'])['pts'].sum()
```
